Load the "diamond" dataset from Seaborn.

```python
import seaborn as sns

diamonds_df = sns.load_dataset('diamonds')

print(diamonds_df.head())
```

Create a boxplot of the "price" variable in the "diamonds" dataset. Set the orientation to
horizontal

```python
sns.boxplot(data=diamonds_df, y='price', orient='h')

sns.xlabel('Price')
sns.title('Boxplot of Price in Diamonds Dataset')

sns.show()
```

Create a boxplot of the "price" variable in the "diamonds" dataset, grouped by the "cut"
variable.

```python
sns.boxplot(data=diamonds_df, x='cut', y='price')

sns.xlabel('Cut')
sns.ylabel('Price')
sns.title('Boxplot of Price by Cut in Diamonds Dataset')

sns.show()
```

Create a boxplot of the "price" variable in the "diamonds" dataset, grouped by both the "cut"
and "color" variables.

```python
sns.boxplot(data=diamonds_df, x='cut', y='price', hue='color')

sns.xlabel('Cut')
sns.ylabel('Price')
sns.title('Boxplot of Price by Cut and Color in Diamonds Dataset')

sns.show()
```

Create a boxplot of the "price" variable in the "diamonds" dataset, grouped by the "cut" variable
and with notched boxes.

```python
sns.boxplot(data=diamonds_df, x='cut', y='price', notch=True)

sns.xlabel('Cut')
sns.ylabel('Price')
sns.title('Boxplot of Price by Cut in Diamonds Dataset (with Notched Boxes)')

sns.show()
```
