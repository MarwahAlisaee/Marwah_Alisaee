# Visualization

```python
import matplotlib.pyplot as plt

# read the dataset from csv file:

df = pd.read_csv("Heart_Disease_Prediction.csv")

# define the columns we want to display it in scatter graph:

x = df['Age']
y = df['BP']

# plotting the data using scatter:

plt.scatter(x, y, color='green')

# Title Lable:

plt.title("To see The Relations Between Age & BP")

# X , Y Lable:

plt.ylabel('Blood Pressure')
plt.xlabel('Age')

# display the scatter grahp:

plt.show()
```

### Observation: The scatter plot was used to try and identify a relation between the 'Age' and 'BP', through the plot there is a slight pattern observed with the increase of the 'Age' there are cases with a higher 'BP' as it is shown in the plot, 'Ages' in the range of 40-65 the 'BP' has shown a significant increase from 120( the normal BP rate) - 160.

### Furthermore, data observed from the 'Age' 66-70 the 'BP' shows a decrease from 160- 125 (closer to a normal 'BP' rate), however; it is an inconclusive result that needs more factors taken into consideration.

```python
# seaborn library provides a high-level interface for drawing attractive and informative statistical graphics
import seaborn as sns 


df = pd.read_csv("Heart_Disease_Prediction.csv")

# plotting the data using scatter plot:

sns.scatterplot(data = df, x = 'Age', y = 'BP', hue = 'Sex')

# Title Lable:

plt.title("To see The Relations Between Age & BP")
```

###  Note: 0 = Female , 1 = Male
### Observation: In order to get a clearer indication to the relation between having a higher 'BP' based on the 'Age', the gender ('Sex') is taken into account. In the 'Age' range of 40-65 with 'BP' of 120 - 160 the majorty of the patients are males in other words, the increase of age in males ,increases the 'BP', as a result there is an increased possibility of getting high blood pressure.
### On the other hand, data observed from the 'Age' range of 66-70 the 'BP' shows a decrease from 160- 125  in females, as a result it is concluded that females are at a lower risk of getting higher blood pressure as their ages increase.

```python
plt.rcParams["figure.figsize"] = [10.00, 5.00]
plt.rcParams["figure.autolayout"] = True

fig, axs = plt.subplots(1,2)


sns.boxplot(x='Age', y='Heart Disease', hue ="Sex" , data=df, ax=axs[0]).set(title="Heart Disease presence based on Age")
sns.boxplot(x='BP', y='Heart Disease', hue ="Sex" ,data=df, ax=axs[1]).set(title="Heart Disease presence based on Blood Pressure")
```

### Observation: In the "Heart Disease presence based on Age" plot it is noticed that male patients between the ages of 50-60 have a higher tendency to have the presence of heart diseases as opposed to their female counterpart whom in the same approximate age range have a lower tendency to have presence of heart diseases.   In the "Heart Disease presence based on Blood Pressure", plot it is noticed that female patients that have a BP  between 130-155  have a higher tendency to have the presence of heart diseases as opposed to their male counterpart whom in the BP range of  118 - 140 have a lower tendency to have presence of heart diseases.
