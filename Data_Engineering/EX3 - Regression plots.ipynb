Create a basic linear regression plot of carat vs. price in the diamonds dataset using lmplot. Add
a 95% confidence interval to the plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds_df = sns.load_dataset('diamonds')

sns.lmplot(data=diamonds_df, x='carat', y='price', ci=95)

plt.xlabel('Carat')
plt.ylabel('Price')
plt.title('Linear Regression Plot of Carat vs. Price')

plt.show()
```

Create a custom linear regression plot of carat vs. price in the diamonds dataset using
plt.scatter(), plt.plot(), and np.polyfit(). Fit quadratic and cubic trends to the data and plot them
on the same plot as the scatter plot and the linear regression line.

```python
import numpy as np

diamonds_df = sns.load_dataset('diamonds')

x = diamonds_df['carat']
y = diamonds_df['price']

plt.scatter(x, y, label='Data')

linear_coeffs = np.polyfit(x, y, deg=1)
linear_fit = np.poly1d(linear_coeffs)
plt.plot(x, linear_fit(x), color='r', label='Linear')

quadratic_coeffs = np.polyfit(x, y, deg=2)
quadratic_fit = np.poly1d(quadratic_coeffs)
plt.plot(x, quadratic_fit(x), color='g', label='Quadratic')

cubic_coeffs = np.polyfit(x, y, deg=3)
cubic_fit = np.poly1d(cubic_coeffs)
plt.plot(x, cubic_fit(x), color='b', label='Cubic')

plt.xlabel('Carat')
plt.ylabel('Price')
plt.title('Custom Linear Regression Plot of Carat vs. Price')
plt.legend()

plt.show()
```

Create a scatter plot of carat vs. price in the diamonds dataset using lmplot. Use the cut column
as the hue variable to group the data by diamond cut. Add a title and legend to the plot.

```python
sns.lmplot(data=diamonds_df, x='carat', y='price', hue='cut', scatter_kws={'alpha': 0.5})

plt.title('Scatter Plot of Carat vs. Price by Diamond Cut')
plt.legend(title='Cut')

plt.show()
```
