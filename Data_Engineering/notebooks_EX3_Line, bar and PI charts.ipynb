## (Line, bar and PI charts)


### â€¢ Load the following data from csv file or from dictionary:


```python
import pandas as pd
data = pd.read_csv('Data_1.csv')
data_dict = {
    'Year': [2015, 2016, 2017, 2018, 2019, 2020],
    'Revenue': [1000000, 1200000, 1400000, 1600000, 1800000, 2000000],
    'Profit': [200000, 300000, 400000, 500000, 600000, 700000],
    'Expenses': [800000, 900000, 1000000, 1100000, 1200000, 1300000]
}
data = pd.DataFrame(data_dict)
data
```

### 1. Create a line chart showing the revenue over the years.


```python
import matplotlib.pyplot as plt

plt.plot(data['Year'], data['Revenue'])
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.title('Revenue Over the Years')
plt.show()
```

### 2. Create a bar chart showing the revenue, profit, and expenses for the year 2019.


```python
plt.bar(['Revenue', 'Profit', 'Expenses'], data.loc[data['Year'] == 2019, ['Revenue', 'Profit', 'Expenses']].values[0])
plt.xlabel('Category')
plt.ylabel('Amount')
plt.title('Financials for 2019')
plt.show()
```

### 3. Create a pie chart showing the percentage of revenue, profit, and expenses for the year 2020.


```python
plt.pie(data.loc[data['Year'] == 2020, ['Revenue', 'Profit', 'Expenses']].values[0], labels=['Revenue', 'Profit', 'Expenses'], autopct='%1.1f%%')
plt.title('Financials for 2020')
plt.show()
```

### 4. Create a stacked bar chart showing the revenue, profit, and expenses for all years.


```python
plt.pie(data.loc[data['Year'] == 2020, ['Revenue', 'Profit', 'Expenses']].values[0], labels=['Revenue', 'Profit', 'Expenses'], autopct='%1.1f%%')
plt.title('Financials for 2020')
plt.show()
```

### 5. Create a scatter plot showing the relationship between revenue and profit for all years.


```python
plt.bar(data['Year'], data['Revenue'], label='Revenue')
plt.bar(data['Year'], data['Profit'], bottom=data['Revenue'], label='Profit')
plt.bar(data['Year'], data['Expenses'], bottom=data['Revenue'] + data['Profit'], label='Expenses')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Financials Over the Years')
plt.legend()
plt.show()
```

### 6. Create a line chart showing the percentage change in revenue over the years.


```python
revenue_change = data['Revenue'].pct_change() * 100
plt.plot(data['Year'], revenue_change)
plt.xlabel('Year')
plt.ylabel('Percentage Change')
plt.title('Percentage Change in Revenue Over the Years')
plt.show()
```

### 7. Create a bar chart showing the revenue for each year, with a horizontal line representing the average revenue across all years.


```python
plt.bar(data['Year'], data['Revenue'])
average_revenue = data['Revenue'].mean()
plt.axhline(average_revenue, color='r', linestyle='--', label='Average Revenue')
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.title('Revenue for Each Year')
plt.legend()
plt.show()
```

### 8. Create a pie chart showing the percentage of revenue for each year.


```python
revenue_percentages = (data['Revenue'] / data['Revenue'].sum()) * 100
plt.pie(revenue_percentages, labels=data['Year'], autopct='%1.1f%%')
plt.title('Percentage of Revenue for Each Year')
plt.show()
```

### 9. Create a line chart showing the revenue and expenses over the years, with two lines


```python
plt.plot(data['Year'], data['Revenue'], label='Revenue')
plt.plot(data['Year'], data['Expenses'], label='Expenses')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Revenue and Expenses Over the Years')
plt.legend()
plt.show()
```

### 10. Create a bar chart showing the revenue and expenses over the years

```python
plt.bar(data['Year'], data['Revenue'], label='Revenue')
plt.bar(data['Year'], data['Expenses'], label='Expenses')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Revenue and Expenses Over the Years')
plt.legend()
plt.show()
```
