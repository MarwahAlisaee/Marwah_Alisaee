```python
ax = plt.subplot(1,1,1)
x = pd.Series(df1.index,dtype='category')
ax.scatter(x,y=df1['Sleep Duration'].to_numpy(),marker='^',color='green')
ax.scatter(x,y=df1['Quality of Sleep'].to_numpy(),color='blue')
plt.xticks(rotation=90)
plt.legend(['Sleep Duration','Quality of Sleep'])
plt.ylim(0,12)
plt.title('Variation of Sleep Quality with Profession')
plt.show()
```

#### above plot displays the relationship  of sleep quality and sleep duration  across different professions , is created using the sleep quality,duration values on the y-axis and the corresponding professions on the x-axis. Each data point represents an individual's sleep quality,duration and their respective profession.

#### we can conclude below points:
#### Sales Representatives have the poorest Sleep Quality and sleep duration, while Engineers have the highest sleep quality and duration

#### Looking the scatter plots  there is a positive relationship between sleep quality and sleep time, and the strength of this relationship is evident in the Manager and software engineer job, as the variation between sleep quality and sleep time values is very small.As for the sales  Representatives and scientist shows that there is a difference between the value of sleep quality and sleep time, and this may be due to other influences that caused a decrease in sleep quality, although the duration of sleep is appropriate.

```python
df2 = sleep_data.groupby('Quality of Sleep')[['Physical Activity Level','Stress Level']].mean()
df2
```

```python
ax = plt.subplot(1,1,1)
x = df2.index
ax.scatter(x,df2['Physical Activity Level'].to_numpy()/10,marker='o',color='green')
ax.scatter(x,df2['Stress Level'].to_numpy(),marker='^',color='red')
ax.set_xlabel('Quality of Sleep-->')
plt.ylim(0,10)
plt.title('Physical Activity & Stress Level vs Quality of Sleep')
plt.legend(['Physical Activity Level','Stress Level'])
plt.show()
```

####  the scatter plot is created using physical activity values and stress level values  on the y-axis, quality of sleep values on the x-axis.
#### Upon examining the scatter plot, we can observe the following :

#### Physical Activity and Stress Levels:

#### There is an inverse correlation between physical activity level and stress level. As physical activity level increases, stress level tends to decrease.
#### This relationship implies that individuals who engage in higher physical activity levels tend to experience lower levels of stress.

#### Physical Activity and Quality of Sleep:

#### There is a positive correlation between physical activity level and the quality of sleep. As physical activity level increases, the quality of sleep tends to improve.This indicates that individuals who engage in higher levels of physical activity are more likely to experience better sleep quality.

#### Stress Levels and Quality of Sleep:

#### There is a negative correlation between stress level and the quality of sleep. As stress level increases, the quality of sleep tends to decline.This suggests that individuals who experience higher levels of stress are more likely to have poorer sleep quality.

#### In summary, the scatter plot analysis reveals that physical activity level and stress level are inversely correlated. Physical activity level shows a positive correlation with the quality of sleep, while stress level demonstrates a negative correlation with sleep quality. These findings highlight the importance of physical activity in reducing stress and improving sleep quality.

```python
sleep_data['Sleep Disorder'].value_counts(normalize=True).sort_values(ascending=False)\
.plot(kind='pie', title='sleep Disorder percentage in the data',explode=[0.05,0.06,0.07],
labels=['None','sleep','Insomnia'],
colors=['#00FFBF','#819FF7','#E2A9F3'],
autopct='%1.2f%%',shadow=True)
```

#### The pie chart displays the following proportions of sleep disorders in a given dataset:

#### None: 58.56%
#### Sleep Apnea: 20.86%
#### Insomnia: 20.59%
#### Interpreting the pie chart:
#### None: The largest segment in the pie chart represents individuals who do not have any diagnosed sleep disorder, accounting for 59% of the total. This suggests that a majority of individuals in the dataset do not have a specific sleep disorder diagnosis.

#### Sleep Apnea: The second largest segment represents individuals diagnosed with sleep apnea, making up 21% of the total. Sleep apnea is a condition characterized by pauses in breathing during sleep.

#### Insomnia: The third largest segment represents individuals diagnosed with insomnia, accounting for 20% of the total. Insomnia refers to difficulties falling asleep or staying asleep.

#### Overall, the pie chart provides an overview of the distribution of sleep disorders within the dataset. It shows that a significant portion of individuals have no diagnosed sleep disorder, while sleep apnea and insomnia are present in a relatively smaller percentage of the population.










```python
ordered_labels = sleep_data.groupby('Sleep Disorder').Age.median().sort_values(ascending=False).index.values

fig, ax = plt.subplots(figsize=(12,6))
sns.boxplot(data=sleep_data, x="Sleep Disorder", y="Age", order=ordered_labels, ax=ax)
plt.xticks(rotation = 80)
plt.xlabel('Sleep Disorder')
plt.ylabel('Age')
plt.title('Sleep Disorders')
plt.show()
```
