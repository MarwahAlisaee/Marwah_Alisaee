```python
df1=pd.read_csv("Sleep_health_.csv")
df1
```

```python
ax = plt.subplot(1,1,1)
x = pd.Series(df1.index,dtype='category')
ax.scatter(x,y=df1['Sleep Duration'].to_numpy(),marker='^',color='green')
ax.scatter(x,y=df1['Quality of Sleep'].to_numpy(),color='blue')
plt.xticks(rotation=90)
plt.legend(['Sleep Duration','Quality of Sleep'])
plt.ylim(0,12)
plt.title('Variation of Sleep Quality with Profession')
plt.show()
```

Each data point represents an individual's sleep quality,duration and their respective profession.

#### we can conclude below points:
#### Sales Representatives have the poorest Sleep Quality and sleep duration, while Engineers have the highest sleep quality and duration

#### Looking the scatter plots  there is a positive relationship between sleep quality and sleep time, and the strength of this relationship is evident in the Manager and software engineer job, as the variation between sleep quality and sleep time values is very small.As for the sales  Representatives and scientist shows that there is a difference between the value of sleep quality and sleep time, and this may be due to other influences that caused a decrease in sleep quality, although the duration of sleep is appropriate.


```python
ax = plt.subplot(1,1,1)
x = df1.index
ax.scatter(x,df1['Physical Activity Level'].to_numpy()/10,marker='o',color='green')
ax.scatter(x,df1['Stress Level'].to_numpy(),marker='^',color='red')
ax.set_xlabel('Quality of Sleep-->')
plt.ylim(0,10)
plt.title('Physical Activity & Stress Level vs Quality of Sleep')
plt.legend(['Physical Activity Level','Stress Level'])
plt.show()
```

####  the scatter plot is created using physical activity values and stress level values  on the y-axis, quality of sleep values on the x-axis.
#### Upon examining the scatter plot, we can observe the following :

#### Physical Activity and Stress Levels:

#### There is an inverse correlation between physical activity level and stress level. As physical activity level increases, stress level tends to decrease.
#### This relationship implies that individuals who engage in higher physical activity levels tend to experience lower levels of stress.

#### Physical Activity and Quality of Sleep:

#### There is a positive correlation between physical activity level and the quality of sleep. As physical activity level increases, the quality of sleep tends to improve.This indicates that individuals who engage in higher levels of physical activity are more likely to experience better sleep quality.

#### Stress Levels and Quality of Sleep:

#### There is a negative correlation between stress level and the quality of sleep. As stress level increases, the quality of sleep tends to decline.This suggests that individuals who experience higher levels of stress are more likely to have poorer sleep quality.

#### In summary, the scatter plot analysis reveals that physical activity level and stress level are inversely correlated. Physical activity level shows a positive correlation with the quality of sleep, while stress level demonstrates a negative correlation with sleep quality. These findings highlight the importance of physical activity in reducing stress and improving sleep quality.

### bar chart ----->'Occupation vs. Stress'

#### creating a stacked bar chart using the pd.crosstab() function to calculate the cross-tabulation between the 'Occupation' and 'Stress Level' columns in the DataFrame 'df'. Here's the code with the necessary imports and adjustments:

```python
# Look closely to stress level by profession

occupation_stress = pd.crosstab(df1['Occupation'], df1['Stress Level'])
occupation_stress.plot(kind='bar', stacked=True)

plt.xlabel('Occupation')
plt.ylabel('Stress')
plt.title('Occupation vs. Stress')

plt.xticks(rotation=80)
plt.show()
```

##### The bar chart displays the relationship between different occupations and their corresponding stress levels. The occupations are listed on the horizontal axis, while the stress levels are represented by the vertical axis. The stress levels range from 1 to 8.

##### From the chart, we can observe that doctors have the highest stress level, consistently rated at 8. Nurses and sales representatives also experience significant stress, with stress levels ranging from 6 to 8. Software engineers and teachers have moderate stress levels, mostly rated between 5 and 7. Engineers, accountants, scientists, lawyers, and managers generally have lower stress levels, ranging from 3 to 6. Salespersons and managers exhibit varying stress levels, with some individuals reporting high stress levels while others report lower levels. Overall, the chart provides an overview of the stress levels associated with different occupations, highlighting the varying degrees of stress individuals may experience in their professional lives.

### Box plot of Sleep Duration by BMI Category

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df1, x='BMI Category', y='Sleep Duration')
plt.title("Sleep Duration by BMI Category")
plt.xlabel("BMI Category")
plt.ylabel("Sleep Duration")
plt.show()
```

##### The box plot illustrates the distribution of sleep duration based on BMI category. The sleep duration is represented on the vertical axis, while the BMI category is displayed on the horizontal axis.

##### The box plot reveals the variation in sleep duration across different BMI categories. There are three BMI categories represented: "Normal Weight," "Overweight," and "Obese."

##### For the "Normal Weight" category, the box plot shows a relatively narrow box, indicating a smaller range of sleep duration values. The median sleep duration appears to be around 6.2 hours, with the box extending from approximately 6 hours to 7.9 hours. There are a few outliers on both the lower and higher ends of the sleep duration spectrum.

##### In the "Overweight" category, the box plot displays a similar range of sleep duration values. The median sleep duration is around 6.6 hours, with the box spanning from approximately 6 hours to 7.7 hours. There are also a few outliers present.

##### The "Obese" category exhibits a slightly wider range of sleep duration. The median sleep duration is around 6.2 hours, and the box extends from roughly 5.9 hours to 6.3 hours. No outliers are visible in this category.

##### Overall, the box plot suggests that there may be some differences in sleep duration based on BMI category. However, it is important to note that this representation is based on a sample dataset, and individual variations and other factors may influence sleep duration. Further analysis and a larger dataset would be necessary to draw more conclusive insights about the relationship between BMI category and sleep duration.
