Load the "tips" dataset from Seaborn.

```python
import seaborn as sns

tips_df = sns.load_dataset('tips')

print(tips_df.head())
```

Create a histogram of the tip amounts using Matplotlib

```python
import matplotlib.pyplot as plt

plt.hist(tips_df['tip'], bins=10, edgecolor='black')

plt.xlabel('Tip Amount')
plt.ylabel('Frequency')
plt.title('Histogram of Tip Amounts')

plt.show()
```

Create a kdeplot of the total bill amounts using Seaborn

```python
sns.kdeplot(data=tips_df['total_bill'])

sns.xlabel('Total Bill Amount')
sns.ylabel('Density')
sns.title('KDE Plot of Total Bill Amounts')

sns.show()
```

Create a jointplot of the total bill and tip amounts using Seaborn.

```python
sns.jointplot(data=tips_df, x='total_bill', y='tip')

sns.xlabel('Total Bill Amount')
sns.ylabel('Tip Amount')
sns.title('Joint Plot of Total Bill and Tip Amounts')

sns.show()
```

Create a pairplot of the tips dataset using Seaborn.
